```python
from tkinter import *
from PIL import ImageTk,Image

root = Tk()
root.title('Image Viewer')
root.iconbitmap('Capture1.ico')



my_img1=ImageTk.PhotoImage(Image.open('IMG-20201103-WA0032.jpg'))
my_img2=ImageTk.PhotoImage(Image.open('Screenshot_20201128_214253.jpg'))
my_img3=ImageTk.PhotoImage(Image.open('Screenshot_20201128_214312.jpg'))
my_img4=ImageTk.PhotoImage(Image.open('Screenshot_20201128_214420.jpg'))
my_img5=ImageTk.PhotoImage(Image.open('IMG-20201104-WA0014.jpg'))
my_img6=ImageTk.PhotoImage(Image.open('Screenshot_20201128_214405.jpg'))
my_img7=ImageTk.PhotoImage(Image.open('Screenshot_20201128_214332.jpg'))

image_list = [my_img1, my_img2, my_img3, my_img4, my_img5,my_img6,my_img7]



my_label = Label(image=my_img1)
my_label.grid(row=1, column=0, columnspan=3)
																	
def forward(image_number):
	global my_label
	global button_forward
	global button_back

	my_label.grid_forget()
	my_label = Label(image=image_list[image_number-1])
	button_forward = Button(root, text="Next Image", command=lambda: forward(image_number+1))
	button_back = Button(root, text="Previous Image", command=lambda: back(image_number-1))
	
	if image_number == 7:
		button_forward = Button(root, text="Next Image", state=DISABLED)

	my_label.grid(row=1, column=0, columnspan=3)
	button_back.grid(row=0, column=0)
	button_forward.grid(row=0, column=2)

def back(image_number):
	global my_label
	global button_forward
	global button_back

	my_label.grid_forget()
	my_label = Label(image=image_list[image_number-1])
	button_forward = Button(root, text="Next Image", command=lambda: forward(image_number+1))
	button_back = Button(root, text="Previous Image", command=lambda: back(image_number-1))

	if image_number == 1:
		button_back = Button(root, text="Previous Image", state=DISABLED)

	my_label.grid(row=1, column=0, columnspan=3)
	button_back.grid(row=0, column=0)
	button_forward.grid(row=0, column=2)



button_back = Button(root, text="Previous Image", command=back, state=DISABLED)
button_exit = Button(root, text="Exit Program", command=root.destroy)
button_forward = Button(root, text="Next Image", command=lambda: forward(2))


button_back.grid(row=0, column=0)
button_exit.grid(row=0, column=1)
button_forward.grid(row=0, column=2)

root.mainloop()
```
